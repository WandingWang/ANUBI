```python
import os

def add_ter_to_pdb(pdb_file_name):
    temp_file_name = f"{pdb_file_name}_temp"  # 临时文件名

    with open(pdb_file_name, 'r') as f:
        lines = f.readlines()

    new_lines = []
    prev_chain_id = None  # 用于跟踪前一个链 ID
    num_lines = len(lines)

    for i, line in enumerate(lines):
        # 保留所有不是 ATOM 的行
        if not line.startswith("ATOM"):
            new_lines.append(line)
            continue

        # 只处理 ATOM 行
        current_chain_id = line[21]  # 提取当前行的链 ID (第22列)
        
        # 添加当前 ATOM 行
        new_lines.append(line)

        # 如果上一条链 ID 不是空，并且后面没有链 ID 数据或链 ID 不一样，则加 TER
        if prev_chain_id is not None:
            # 检查当前 ATOM 行是否是当前链的最后一行，若是则添加 TER
            if (i == num_lines - 1 or
                (lines[i + 1].startswith("ATOM") and lines[i + 1][21] != current_chain_id) or
                not lines[i + 1].startswith("ATOM")):
                # 如果上一行已经有 TER，就不再添加
                if new_lines[-1].startswith("TER"):
                    pass  # 不添加 TER，因为已经存在了
                else:
                    new_lines.append("TER\n")

        # 更新前一个链 ID
        prev_chain_id = current_chain_id

    # 最后一行的链结束没有 TER 时，手动添加
    if prev_chain_id is not None and not new_lines[-1].startswith("TER"):
        new_lines.append("TER\n")

    # 写入修改后的内容到临时文件
    with open(temp_file_name, 'w') as f:
        f.writelines(new_lines)

    # 用临时文件替换原文件
    os.rename(temp_file_name, pdb_file_name)
```
